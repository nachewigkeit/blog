```python
from transformers import AutoModelForSequenceClassification, AutoTokenizer
from bertviz import neuron_view, model_view
import torch
```

```python
model = AutoModelForSequenceClassification.from_pretrained('./weight/base', output_attentions=True).cuda()
tokenizer = AutoTokenizer.from_pretrained('uer/roberta-base-finetuned-dianping-chinese')
```

```python
inputs = tokenizer("这是在携程能找到的、汕尾最好的酒店。没办法")
token_type_ids = torch.tensor([inputs['token_type_ids']]).cuda()
input_ids = torch.tensor([inputs['input_ids']]).cuda()
attention = model(input_ids, token_type_ids=token_type_ids)[-1]
input_id_list = input_ids[0].tolist() # Batch index 0
tokens = tokenizer.convert_ids_to_tokens(input_id_list)
```

```python
print(len(attention))
print(attention[11][0,0,0,:])
```

```python
# 用BertViz可视化
neuron_view.show(attention, tokens)
```

```python
model_view(attention, tokens)
```
